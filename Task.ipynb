### Auction Setup

```python
import numpy as np
```

```python
class item: #Item object is created
    def __init__(self, item_no, reserve, description):
        self.item_no = item_no
        self.reserve = reserve
        self.description = description
        self.bids = 0
        self.is_sold = False
        self.bidder_ID = None
        self.highest_bid = 0
        self.total_cost = 0
```

```python
items = []

while True:
    
    try: #Try catch blocks to ensure correct datatype in inputted
        choice = int(input("Please enter 0 to quit, 1 to input item: ")) #Asking user to quit or continue inputting data
        if choice == 1:
            item_no = int(input("Please enter item number: "))
            for i in items: #Checking if item number is already in use
                if i.item_no == item_no:
                    raise ValueError()
            reserve = int(input("Please enter reserve price of item: "))
            description = input("Please enter description of item: ")
        elif choice == 0:
            if len(items) < 4: #Items should be more than 10, setting to 4 for ease
                print("No items shall be saved as number of items is less than 10.")
                items.clear()
            break
        else:
            print("Wrong input, please try again.")
            continue
    except:
        print("Error: Wrong datatype, Reenter data")
        continue
    
    obj = item(item_no, reserve, description) #Object of item is created and saved
    items.append(obj)
    
```

### Auction Setup

```python
while True:
    try:
        choice = int(input("Please enter 0 to end bidding, 1 to continue bidding: "))
        if choice == 1:
            ID = int(input("Please enter your buyer ID: "))
            item_no = int(input("Please enter which item number you'd like to view: "))
            is_found = False
            for i in items:
                if i.item_no == item_no:
                    is_found = True
                    print("Following is the item data:\nItem Number : "+str(i.item_no)+"\nReserve price: "+str(i.reserve)+
                         "\nDescription : "+i.description+"\nNumber of bids : "+
                          str(i.bids)+"\nCurrent Highest Bid : "+str(i.highest_bid))
                    will_bid = input("Would you like to bid for the item? (y/n)")
                    if will_bid == "y":
                        bid = int(input("Please enter your bid value: "))  # Convert input to int
                        if i.highest_bid < bid:
                            i.highest_bid = bid
                            i.bids += 1
                            break
                        else:
                            print("Bid too low.")
                            break
                    elif will_bid == "n":
                        print("Acknowledged.")
                        break
                    else:
                        raise ValueError()
            if not is_found:
                print("Item does not exist.")
                continue
        elif choice == 0:
            break
        else:
            print("Wrong input, please try again.")
            continue
        
    except ValueError:
        print("Error: Invalid input for bid value. Please enter a valid integer.")
        continue
```

### Auction Finalization

```python
print("Sold Items are as Follows: \n")
for i in items:
    if i.highest_bid > i.reserve:
        i.is_sold = True
        total_cost = i.highest_bid + (i.highest_bid *0.1)
        print("Item Number : "+str(i.item_no)+"\nReserve price: "+str(i.reserve)+
            "\nDescription : "+i.description+"\nNumber of bids : "+
            str(i.bids)+"\nCurrent Highest Bid : "+str(i.highest_bid)+"\nTotal Cost : "+str(i.total_cost)+"\n\n\n")
        
        
print("Unsold Items are as Follows: \n")
for i in items:
    if i.highest_bid < i.reserve and i.bids != 0:
        print("Item Number : "+str(i.item_no)+"\nReserve price: "+str(i.reserve)+
            "\nDescription : "+i.description+"\nNumber of bids : "+
            str(i.bids)+"\n\n\n")


print("Items with no bids are as Follows:\n")       
for i in items:
    if i.highest_bid < i.reserve and i.bids ==0:
        print("Item Number : "+str(i.item_no)+"\nReserve price: "+str(i.reserve)+
            "\nDescription : "+i.description+"\nNumber of bids : "+
            str(i.bids)+"\n\n\n")
```
